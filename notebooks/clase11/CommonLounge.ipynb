# Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
x
```

```python
S, C = np.sin(x), np.cos(x)
```

```python
plt.figure()
plt.plot(x, S, '-')
plt.plot(x, C, '--')
```

```python
fig = plt.figure(figsize=(10,6), dpi=80)
# Definimos colores y ancho de lneas, junto con etiquetas
plt.plot(x, C, color="blue", linewidth=2.5, linestyle="--", label="cosine")
plt.plot(x, S, color="green", linewidth=2.5, linestyle="-", label="sine")
# Lmite del etiquetas
plt.xlim(-4.0, 4.0)
# Cuantos nmeros en el eje X
plt.xticks(np.linspace(-4, 4, 9, endpoint=True)) 
# Lmite en Y
plt.ylim(-1.0, 1.0)
# Cuantos nmeros en Y
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))
# Donde colocar el ttulo
plt.legend(loc='upper left', frameon=False)
# El ttulo
plt.title("Graph of wave movement with Sine and Cosine functions")
# Etiquetas de los ejees
plt.xlabel("Time, t")
plt.ylabel("Position, x")
# Colocamos una grilla
plt.grid(color='b', linestyle='-', linewidth=0.1)
# Obtenemos los ejes
ax = plt.gca()
# Colocamos los ejes en el centro
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
# Eliminamos los laterales
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
```

```python
# Guarda la grfica
fig.savefig('my_figure.png')
```
