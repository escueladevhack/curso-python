```python
data = pd.read_csv('UKretail.csv', encoding="ISO-8859-1")
```

```python
data
```

```python
data.shape
```

```python
# Graficando directamente desde pandas
data.plot(x="InvoiceDate", y="UnitPrice", kind="line")
```

```python
# Algunos filtros
positive = data[data["UnitPrice"] > 0]
positive.plot(x="InvoiceDate", y="UnitPrice", kind="line")
```

```python
# Agregaciones
objects_sold = data.groupby("StockCode")["UnitPrice"].sum().reset_index()
objects_sold
```

```python
# Mas filtros
just_DOT = data[data["StockCode"] == "DOT"]
just_DOT
```

```python
# Historiogramas
just_DOT["UnitPrice"].plot(kind="hist")
```

```python
just_DOT["UnitPrice"].hist()
```
